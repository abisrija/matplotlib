# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, marker='_')
plt.show()
```

# marker|line|color

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([12,10,20,30,66])
plt.plot(a,'*:g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([10,90,60,40,77])
b=np.array([98,76,45,23,54])
plt.plot(a,b,marker='o',ms=20,mec='r')#marker size,marker color
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([45,64,76,87,98])
"""ls= dashed,dotted,solid,dashdot 
    c or color='r','b','g' etc
    linewidth
"""
plt.plot(a,linestyle='dotted',c='hotpink',lw=20)
plt.show()
```

```python
a=([12,23,34,45,67])
b=([90,78,56,45,34])
plt.plot(a,b)
plt.title("health is wealth",loc='left')
plt.xlabel("health")
plt.ylabel("wealth")
plt.show()
```

```python
a=([12,23,34,45,67])
b=([90,78,56,45,34])
plt.plot(a,b)
plt.title("health is wealth",loc='left')
plt.xlabel("health")
plt.ylabel("wealth")
plt.grid()
plt.show()
```
